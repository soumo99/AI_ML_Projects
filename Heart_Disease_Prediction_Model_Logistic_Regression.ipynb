```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/Users/soumobratamanna/Documents/AI_ML PROJECTS/Heart_Disease_Prediction.csv")
```

```python
#Encoding Presence  and Absence into 1 and 0
data['Heart Disease'] = data['Heart Disease'].map({'Presence':1,'Absence':0})
```

```python
data['Heart Disease'].unique()
```

```python
data.head(10)
```

```python
data.describe()
```

```python
#checking the number of PAtients having heart diseases 
data['Heart Disease'].value_counts()
```

```python
#Splitting the datas between independent and dependent variable of X and Y
X = data.iloc[:,:-1]
Y = data.iloc[:,-1]
```

```python
X
```

```python
Y
```

```python
#Dividing data between training and testing data
#stratify = y is used to divide the the columns of Y evenly 
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify = Y,random_state=2)
```

```python
X.shape
```

```python
Y.shape
```

```python
X_train
```

```python
X_test
```

```python
#Training the mdoel
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#Evaluation metrics 
```

```python
#Checking the accuracy score of the training dataset
```

```python
X_train_predictions = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predictions,Y_train)
```

```python
training_data_accuracy
```

```python
#Checking the accuracy score of the testing dataset
```

```python
X_test_predictions = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_predictions,Y_test)
```

```python
testing_data_accuracy
```

```python
#Imp Note - If the accuracy score of training data is way too large and the accuracy score of testing data is way too small then the model is overfitted
#The accuracy score between training and testing data should be near about or almost similar or close to each other .
```

```python
# BUILDING A PREDICTIVE SYSTEM - 1 means heart disease present and 0 means not present 
```

```python
input_data = (70,1,4,130,322,0,2,109,0,2.4,2,3,3)

#changing to numpy array beacuse we have to reshape it and its easy to reshape numpy array than a tuples 
input_data_as_numpy_array = np.asarray(input_data)

#Reshaping because we need to predict the value of 1 data row
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

predictions = model.predict(input_data_reshaped)

print(predictions)

if(predictions[0] == 0):
    print("The person is healthy heart disease : result negative")
else:
    print("The person is not healthy heart disease : result positive")


```
